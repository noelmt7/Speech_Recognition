(1) Generate and plot the following signals.
    
    (a) A unit step function.
    (b) A unit impulse function.
    (c) A ramp function.
    (d) An exponential signal (decaying and growing).
    (e) A sinusoidal signal.

```python
import numpy as np
import matplotlib.pyplot as plt

# (a) Unit step function
t = np.linspace(-10, 10, 1000)
u = np.heaviside(t, 1)

# (b) Unit impulse function
unit_impulse = np.zeros_like(t)
unit_impulse[np.where(np.isclose(t, 0))] = 1

# (c) Ramp function
ramp = np.maximum   (0, t)

# (d) Exponential signal (decaying and growing)
exponential_decaying = np.exp(-t)
exponential_growing = np.exp(t)

# (e) Sinusoidal signal
frequency = 2  # Hz
sinusoidal = np.sin(2 * np.pi * frequency * t)

# Plotting the signals
plt.figure(figsize=(12, 8))

plt.subplot(5, 1, 1)
plt.plot(t, u)
plt.title('Unit Step Function')
plt.grid(True)

plt.subplot(5, 1, 2)
plt.plot(t, unit_impulse)
plt.title('Unit Impulse Function')
plt.grid(True)

plt.subplot(5, 1, 3)
plt.plot(t, ramp)
plt.title('Ramp Function')
plt.grid(True)  

plt.subplot(5, 1, 4)
plt.plot(t, exponential_decaying)
plt.plot(t, exponential_growing)
plt.title('Exponential Signal')
plt.grid(True)

plt.subplot(5, 1, 5)
plt.plot(t, sinusoidal)
plt.title('Sinusoidal Signal')
plt.grid(True)

plt.tight_layout()
plt.show()
```

(2) You are asked to visualize the effects of sampling and reconstructing a continuous-
time signal.

    (a) Generate a continuous sinusoidal signal.
    (b) Sample the signal at different rates (Nyquist rate, above, and below Nyquist).
    (c) Reconstruct the sampled signal and observe the aliasing effect when un-dersampled.
    (d) Plot the continuous signal, sampled points, and the reconstructed signal.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
f = 5  # Frequency of the sinusoidal signal (Hz)
T = 1  # Duration of the signal (seconds)
Fs_cont = 1000  # Sampling frequency for continuous signal (to simulate continuous-time)

# Time vector for continuous signal
t_cont = np.linspace(0, T, int(Fs_cont * T), endpoint=False)

# Continuous sinusoidal signal
signal_cont = np.sin(2 * np.pi * f * t_cont)

# Sampling rates
Fs_nyquist = 2 * f  # Nyquist rate
Fs_above = 4 * f    # Above Nyquist rate
Fs_below = f        # Below Nyquist rate (undersampling)

# Time vectors for sampling
t_nyquist = np.linspace(0, T, int(Fs_nyquist * T), endpoint=False)
t_above = np.linspace(0, T, int(Fs_above * T), endpoint=False)
t_below = np.linspace(0, T, int(Fs_below * T), endpoint=False)

# Sampled signals
signal_nyquist = np.sin(2 * np.pi * f * t_nyquist)
signal_above = np.sin(2 * np.pi * f * t_above)
signal_below = np.sin(2 * np.pi * f * t_below)

# Plotting
plt.figure(figsize=(15, 10))

# (a) Continuous signal
plt.subplot(3, 1, 1)
plt.plot(t_cont, signal_cont, label='Continuous Signal', color='blue')
plt.title('Continuous Sinusoidal Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

# (b) Sampled signals at different rates
# Nyquist rate
plt.subplot(3, 1, 2)
plt.plot(t_cont, signal_cont, label='Continuous Signal', color='blue', alpha=0.7)
plt.stem(t_nyquist, signal_nyquist, linefmt='r-', markerfmt='ro', basefmt='r-', label='Nyquist Sampling')
plt.title('Sampling at Nyquist Rate')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Above and below Nyquist
plt.subplot(3, 1, 3)
plt.plot(t_cont, signal_cont, label='Continuous Signal', color='blue', alpha=0.7)
plt.stem(t_above, signal_above, linefmt='g-', markerfmt='go', basefmt='g-', label='Above Nyquist Sampling')
plt.stem(t_below, signal_below, linefmt='m-', markerfmt='mo', basefmt='m-', label='Below Nyquist Sampling (Aliasing)')
plt.title('Sampling at Above and Below Nyquist Rates')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
